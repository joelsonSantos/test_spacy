```python
import spacy
import pandas as pd
import os
```

```python
data = []
with open("diario.txt", "r") as f:
    for line in f.readlines():
        data.append(line.replace('\n', ''))
# Tia Dorinha não curtiu esta linha abaixo (nem as acima hehe).
dataset = pd.DataFrame({'corpus': list(set(data))})
```

```python
dataset.head()
```

```python
# carrega modelo em portugues
nlp = spacy.load('pt_core_news_sm')
```

```python
text = "\n".join(dataset["corpus"])
doc = nlp(text)
```

```python
def classification(doc):
    result = []
    for entity in doc.ents:
        result.append((entity.text, entity.label_))
    return result
```

```python
classification(doc)
```
